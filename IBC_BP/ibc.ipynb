```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)
pd.set_option("display.max_colwidth", 120)

DATA_PATH = Path("cosmoshub_ibc_2500.csv")
if not DATA_PATH.exists():
    # fallback, když notebook poběží jinde
    DATA_PATH = Path("/mnt/data/cosmoshub_ibc_2500.csv")

print("Using file:", DATA_PATH)

UATOM_PER_ATOM = 1_000_00
ATOM_USD_PRICE = 1.8167  # nastavil jsme cenu k 3.3. 2026 ve 23:59 h, protože to odopovídalo době, kdy jsem získaval data


# Načtení dat
df = pd.read_csv(DATA_PATH)
print("Shape:", df.shape)
display(df.head())


# Základní info o datech
df.info()


# Missing values summary
missing_summary = (
    df.isna()
      .sum()
      .rename("missing_count")
      .to_frame()
      .assign(missing_share=lambda x: x["missing_count"] / len(df))
      .sort_values("missing_count", ascending=False)
)

display(missing_summary.head(20))


# Převody datových typů
timestamp_cols = ["send_timestamp", "recv_timestamp", "ack_timestamp"]
for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce", utc=True)

numeric_cols = [
    "send_height", "send_code", "packet_sequence",
    "send_gas_wanted", "send_gas_used", "send_fee_amount",
    "recv_height", "recv_code", "recv_gas_used", "recv_fee_amount",
    "ack_height", "ack_code", "ack_gas_used", "ack_fee_amount",
    "latency_send_to_recv_s", "latency_send_to_ack_s", "latency_recv_to_ack_s",
    "has_recv", "has_ack",
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")


# Odvozené metriky
df["user_cost_raw"] = df["send_fee_amount"].fillna(0)
df["relayer_cost_raw"] = df["recv_fee_amount"].fillna(0) + df["ack_fee_amount"].fillna(0)
df["network_cost_raw"] = df["user_cost_raw"] + df["relayer_cost_raw"]

send_fee_is_uatom = df["send_fee_denom"].eq("uatom") | df["send_fee_amount"].fillna(0).eq(0)
recv_fee_is_uatom = df["recv_fee_denom"].eq("uatom") | df["recv_fee_amount"].fillna(0).eq(0)
ack_fee_is_uatom = df["ack_fee_denom"].eq("uatom") | df["ack_fee_amount"].fillna(0).eq(0)

df["send_fee_atom"] = np.where(send_fee_is_uatom, df["send_fee_amount"] / UATOM_PER_ATOM, np.nan)
df["recv_fee_atom"] = np.where(recv_fee_is_uatom, df["recv_fee_amount"] / UATOM_PER_ATOM, np.nan)
df["ack_fee_atom"] = np.where(ack_fee_is_uatom, df["ack_fee_amount"] / UATOM_PER_ATOM, np.nan)

df["user_cost_atom"] = np.where(send_fee_is_uatom, df["user_cost_raw"] / UATOM_PER_ATOM, np.nan)
df["relayer_cost_atom"] = np.where(
    recv_fee_is_uatom & ack_fee_is_uatom,
    df["relayer_cost_raw"] / UATOM_PER_ATOM,
    np.nan,
)
df["network_cost_atom"] = np.where(
    send_fee_is_uatom & recv_fee_is_uatom & ack_fee_is_uatom,
    df["network_cost_raw"] / UATOM_PER_ATOM,
    np.nan,
)

df["send_fee_usd"] = df["send_fee_atom"] * ATOM_USD_PRICE
df["recv_fee_usd"] = df["recv_fee_atom"] * ATOM_USD_PRICE
df["ack_fee_usd"] = df["ack_fee_atom"] * ATOM_USD_PRICE
df["user_cost_usd"] = df["user_cost_atom"] * ATOM_USD_PRICE
df["relayer_cost_usd"] = df["relayer_cost_atom"] * ATOM_USD_PRICE
df["network_cost_usd"] = df["network_cost_atom"] * ATOM_USD_PRICE

df["packet_status"] = np.select(
    [
        (df["has_recv"] == 1) & (df["has_ack"] == 1),
        (df["has_recv"] == 1) & (df["has_ack"] == 0),
        (df["has_recv"] == 0),
    ],
    [
        "recv_and_ack",
        "recv_without_ack",
        "send_only",
    ],
    default="unknown"
)

df["dst_chain_guess"] = df["dst_chain_guess"].fillna("unknown")

display(df.head())


# Sanity checks pro hlavní analytické sloupce
sanity = df[[
    "latency_send_to_recv_s",
    "latency_send_to_ack_s",
    "latency_recv_to_ack_s",
    "send_fee_amount",
    "recv_fee_amount",
    "ack_fee_amount",
    "has_recv",
    "has_ack",
    "user_cost_raw",
    "relayer_cost_raw",
    "network_cost_raw",
    "user_cost_atom",
    "relayer_cost_atom",
    "network_cost_atom",
    "user_cost_usd",
    "relayer_cost_usd",
    "network_cost_usd",
]].describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.90, 0.95, 0.99]).T

display(sanity)


# Přehled datasetu
overview = pd.DataFrame({
    "metric": [
        "rows",
        "unique destination chains",
        "share with recv",
        "share with ack",
        "share send_only",
        "share recv_and_ack",
    ],
    "value": [
        len(df),
        df["dst_chain_guess"].nunique(),
        df["has_recv"].mean(),
        df["has_ack"].mean(),
        (df["packet_status"] == "send_only").mean(),
        (df["packet_status"] == "recv_and_ack").mean(),
    ]
})

display(overview)


# Packet status distribution
status_counts = (
    df["packet_status"]
      .value_counts(dropna=False)
      .rename_axis("packet_status")
      .reset_index(name="count")
)
status_counts["share"] = status_counts["count"] / len(df)
display(status_counts)


plt.figure(figsize=(8, 4))
plt.bar(status_counts["packet_status"], status_counts["count"])
plt.title("IBC packet status distribution")
plt.xlabel("Packet status")
plt.ylabel("Count")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()


# Latency summary
latency_summary = (
    df[[
        "latency_send_to_recv_s",
        "latency_send_to_ack_s",
        "latency_recv_to_ack_s",
    ]]
    .describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.90, 0.95, 0.99])
    .T
)

display(latency_summary)


# Histogramy latencí
latency_cols = [
    "latency_send_to_recv_s",
    "latency_send_to_ack_s",
    "latency_recv_to_ack_s",
]

for col in latency_cols:
    series = df[col].dropna()
    plt.figure(figsize=(8, 4))
    plt.hist(series, bins=50)
    plt.title(f"Distribution of {col}")
    plt.xlabel("Seconds")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()


# Log histogramy latencí (užitečné při skewed distribuci)
for col in latency_cols:
    series = df[col].dropna()
    positive = series[series > 0]
    if len(positive) == 0:
        continue

    plt.figure(figsize=(8, 4))
    plt.hist(np.log10(positive), bins=50)
    plt.title(f"Log10 distribution of {col}")
    plt.xlabel("log10(seconds)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()


# Top chains podle počtu packetů
top_chains = df["dst_chain_guess"].value_counts().head(10).index.tolist()
df_top = df[df["dst_chain_guess"].isin(top_chains)].copy()

chain_latency = (
    df_top.groupby("dst_chain_guess")
         .agg(
             packets=("packet_sequence", "count"),
             avg_send_to_recv=("latency_send_to_recv_s", "mean"),
             median_send_to_recv=("latency_send_to_recv_s", "median"),
             p90_send_to_recv=("latency_send_to_recv_s", lambda s: s.quantile(0.90)),
             p95_send_to_recv=("latency_send_to_recv_s", lambda s: s.quantile(0.95)),
             p99_send_to_recv=("latency_send_to_recv_s", lambda s: s.quantile(0.99)),
             avg_send_to_ack=("latency_send_to_ack_s", "mean"),
             median_send_to_ack=("latency_send_to_ack_s", "median"),
             p90_send_to_ack=("latency_send_to_ack_s", lambda s: s.quantile(0.90)),
             p95_send_to_ack=("latency_send_to_ack_s", lambda s: s.quantile(0.95)),
             p99_send_to_ack=("latency_send_to_ack_s", lambda s: s.quantile(0.99)),
         )
         .sort_values("packets", ascending=False)
         .reset_index()
)

display(chain_latency)


# Boxplot send -> ack latency pro top chainy
box_data = []
labels = []

for chain in top_chains:
    values = df.loc[df["dst_chain_guess"] == chain, "latency_send_to_ack_s"].dropna()
    if len(values) > 0:
        box_data.append(values.values)
        labels.append(chain)

plt.figure(figsize=(10, 5))
plt.boxplot(box_data, tick_labels=labels, showfliers=False)
plt.title("Send-to-ack latency by destination chain (top 10 chains)")
plt.xlabel("Destination chain")
plt.ylabel("Seconds")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()


# Mean vs median latency podle chainu
plot_df = chain_latency.sort_values("avg_send_to_ack", ascending=False)

x = np.arange(len(plot_df))
width = 0.4

plt.figure(figsize=(11, 5))
plt.bar(x - width / 2, plot_df["avg_send_to_ack"], width=width, label="Mean")
plt.bar(x + width / 2, plot_df["median_send_to_ack"], width=width, label="Median")
plt.xticks(x, plot_df["dst_chain_guess"], rotation=30, ha="right")
plt.title("Mean vs median send-to-ack latency by chain")
plt.xlabel("Destination chain")
plt.ylabel("Seconds")
plt.legend()
plt.tight_layout()
plt.show()


# Reliability overall
reliability_overall = pd.DataFrame({
    "metric": ["recv_success_rate", "ack_success_rate"],
    "value": [df["has_recv"].mean(), df["has_ack"].mean()]
})
display(reliability_overall)


# Reliability by chain
reliability_by_chain = (
    df.groupby("dst_chain_guess")
      .agg(
          packets=("packet_sequence", "count"),
          recv_success_rate=("has_recv", "mean"),
          ack_success_rate=("has_ack", "mean"),
      )
      .sort_values(["packets", "ack_success_rate"], ascending=[False, False])
      .reset_index()
)

display(reliability_by_chain.head(20))


# Reliability graf pro top 15 chainů
top_rel = reliability_by_chain.head(15).copy()

x = np.arange(len(top_rel))
width = 0.4

plt.figure(figsize=(12, 5))
plt.bar(x - width / 2, top_rel["recv_success_rate"], width=width, label="Recv success rate")
plt.bar(x + width / 2, top_rel["ack_success_rate"], width=width, label="Ack success rate")
plt.xticks(x, top_rel["dst_chain_guess"], rotation=35, ha="right")
plt.ylim(0, 1.05)
plt.title("Reliability by destination chain (top 15 by packet count)")
plt.xlabel("Destination chain")
plt.ylabel("Rate")
plt.legend()
plt.tight_layout()
plt.show()


# Nejhorší ack success rate při minimálním počtu packetů
reliability_filtered = reliability_by_chain[reliability_by_chain["packets"] >= 10].copy()
worst_ack = reliability_filtered.sort_values("ack_success_rate", ascending=True).head(10)

display(worst_ack)

plt.figure(figsize=(10, 5))
plt.bar(worst_ack["dst_chain_guess"], worst_ack["ack_success_rate"])
plt.ylim(0, 1.05)
plt.title("Lowest ack success rate among chains with at least 10 packets")
plt.xlabel("Destination chain")
plt.ylabel("Ack success rate")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()


# Traffic distribution
traffic_by_chain = (
    df["dst_chain_guess"]
      .value_counts()
      .rename_axis("dst_chain_guess")
      .reset_index(name="packets")
)

traffic_by_chain["share"] = traffic_by_chain["packets"] / len(df)
display(traffic_by_chain.head(20))


top_traffic = traffic_by_chain.head(15)

plt.figure(figsize=(11, 5))
plt.bar(top_traffic["dst_chain_guess"], top_traffic["packets"])
plt.title("IBC traffic distribution by destination chain")
plt.xlabel("Destination chain")
plt.ylabel("Packets")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()


# Cost summary
cost_summary = (
    df[[
        "send_fee_amount",
        "recv_fee_amount",
        "ack_fee_amount",
        "user_cost_raw",
        "relayer_cost_raw",
        "network_cost_raw",
        "user_cost_atom",
        "relayer_cost_atom",
        "network_cost_atom",
        "user_cost_usd",
        "relayer_cost_usd",
        "network_cost_usd",
    ]]
    .describe(percentiles=[0.05, 0.25, 0.5, 0.75, 0.90, 0.95, 0.99])
    .T
)

display(cost_summary)


# Fee denom overview
fee_denoms = {
    "send_fee_denom": df["send_fee_denom"].value_counts(dropna=False).rename_axis("denom").reset_index(name="count"),
    "recv_fee_denom": df["recv_fee_denom"].value_counts(dropna=False).rename_axis("denom").reset_index(name="count"),
    "ack_fee_denom": df["ack_fee_denom"].value_counts(dropna=False).rename_axis("denom").reset_index(name="count"),
}

for name, table in fee_denoms.items():
    print(f"\n{name}")
    display(table.head(15))


# Cost by destination chain
cost_by_chain = (
    df.groupby("dst_chain_guess")
      .agg(
          packets=("packet_sequence", "count"),
          avg_user_cost_raw=("user_cost_raw", "mean"),
          median_user_cost_raw=("user_cost_raw", "median"),
          p90_user_cost_raw=("user_cost_raw", lambda s: s.quantile(0.90)),
          p99_user_cost_raw=("user_cost_raw", lambda s: s.quantile(0.99)),
          avg_user_cost_atom=("user_cost_atom", "mean"),
          median_user_cost_atom=("user_cost_atom", "median"),
          p90_user_cost_atom=("user_cost_atom", lambda s: s.quantile(0.90)),
          p99_user_cost_atom=("user_cost_atom", lambda s: s.quantile(0.99)),
          avg_user_cost_usd=("user_cost_usd", "mean"),
          median_user_cost_usd=("user_cost_usd", "median"),
          p90_user_cost_usd=("user_cost_usd", lambda s: s.quantile(0.90)),
          p99_user_cost_usd=("user_cost_usd", lambda s: s.quantile(0.99)),
          avg_network_cost_raw=("network_cost_raw", "mean"),
          median_network_cost_raw=("network_cost_raw", "median"),
          p90_network_cost_raw=("network_cost_raw", lambda s: s.quantile(0.90)),
          p99_network_cost_raw=("network_cost_raw", lambda s: s.quantile(0.99)),
          avg_network_cost_atom=("network_cost_atom", "mean"),
          median_network_cost_atom=("network_cost_atom", "median"),
          p90_network_cost_atom=("network_cost_atom", lambda s: s.quantile(0.90)),
          p99_network_cost_atom=("network_cost_atom", lambda s: s.quantile(0.99)),
          avg_network_cost_usd=("network_cost_usd", "mean"),
          median_network_cost_usd=("network_cost_usd", "median"),
          p90_network_cost_usd=("network_cost_usd", lambda s: s.quantile(0.90)),
          p99_network_cost_usd=("network_cost_usd", lambda s: s.quantile(0.99)),
      )
      .sort_values("packets", ascending=False)
      .reset_index()
)

display(cost_by_chain.head(20))


# Average raw costs by chain
plot_cost = cost_by_chain.head(12).copy()

x = np.arange(len(plot_cost))
width = 0.4

plt.figure(figsize=(12, 5))
plt.bar(x - width / 2, plot_cost["avg_user_cost_raw"], width=width, label="Avg user cost raw")
plt.bar(x + width / 2, plot_cost["avg_network_cost_raw"], width=width, label="Avg network cost raw")
plt.xticks(x, plot_cost["dst_chain_guess"], rotation=35, ha="right")
plt.title("Average raw costs by destination chain")
plt.xlabel("Destination chain")
plt.ylabel("Raw fee amount")
plt.legend()
plt.tight_layout()
plt.show()


# Median raw costs by chain
plot_cost_med = cost_by_chain.head(12).copy()

x = np.arange(len(plot_cost_med))
width = 0.4

plt.figure(figsize=(12, 5))
plt.bar(x - width / 2, plot_cost_med["median_user_cost_raw"], width=width, label="Median user cost raw")
plt.bar(x + width / 2, plot_cost_med["median_network_cost_raw"], width=width, label="Median network cost raw")
plt.xticks(x, plot_cost_med["dst_chain_guess"], rotation=35, ha="right")
plt.title("Median raw costs by destination chain")
plt.xlabel("Destination chain")
plt.ylabel("Raw fee amount")
plt.legend()
plt.tight_layout()
plt.show()


# Hlavní summary tabulka
summary_table = (
    df.groupby("dst_chain_guess")
      .agg(
          packets=("packet_sequence", "count"),
          recv_success_rate=("has_recv", "mean"),
          ack_success_rate=("has_ack", "mean"),
          avg_send_to_recv_s=("latency_send_to_recv_s", "mean"),
          median_send_to_recv_s=("latency_send_to_recv_s", "median"),
          p90_send_to_recv_s=("latency_send_to_recv_s", lambda s: s.quantile(0.90)),
          p99_send_to_recv_s=("latency_send_to_recv_s", lambda s: s.quantile(0.99)),
          avg_send_to_ack_s=("latency_send_to_ack_s", "mean"),
          median_send_to_ack_s=("latency_send_to_ack_s", "median"),
          p90_send_to_ack_s=("latency_send_to_ack_s", lambda s: s.quantile(0.90)),
          p99_send_to_ack_s=("latency_send_to_ack_s", lambda s: s.quantile(0.99)),
          avg_user_cost_raw=("user_cost_raw", "mean"),
          median_user_cost_raw=("user_cost_raw", "median"),
          p90_user_cost_raw=("user_cost_raw", lambda s: s.quantile(0.90)),
          p99_user_cost_raw=("user_cost_raw", lambda s: s.quantile(0.99)),
          avg_user_cost_atom=("user_cost_atom", "mean"),
          median_user_cost_atom=("user_cost_atom", "median"),
          p90_user_cost_atom=("user_cost_atom", lambda s: s.quantile(0.90)),
          p99_user_cost_atom=("user_cost_atom", lambda s: s.quantile(0.99)),
          avg_user_cost_usd=("user_cost_usd", "mean"),
          median_user_cost_usd=("user_cost_usd", "median"),
          p90_user_cost_usd=("user_cost_usd", lambda s: s.quantile(0.90)),
          p99_user_cost_usd=("user_cost_usd", lambda s: s.quantile(0.99)),
          avg_network_cost_raw=("network_cost_raw", "mean"),
          median_network_cost_raw=("network_cost_raw", "median"),
          p90_network_cost_raw=("network_cost_raw", lambda s: s.quantile(0.90)),
          p99_network_cost_raw=("network_cost_raw", lambda s: s.quantile(0.99)),
          avg_network_cost_atom=("network_cost_atom", "mean"),
          median_network_cost_atom=("network_cost_atom", "median"),
          p90_network_cost_atom=("network_cost_atom", lambda s: s.quantile(0.90)),
          p99_network_cost_atom=("network_cost_atom", lambda s: s.quantile(0.99)),
          avg_network_cost_usd=("network_cost_usd", "mean"),
          median_network_cost_usd=("network_cost_usd", "median"),
          p90_network_cost_usd=("network_cost_usd", lambda s: s.quantile(0.90)),
          p99_network_cost_usd=("network_cost_usd", lambda s: s.quantile(0.99)),
      )
      .sort_values("packets", ascending=False)
      .reset_index()
)

display(summary_table.head(20))


# Focus jen na chainy s alespoň 20 packety
summary_filtered = summary_table[summary_table["packets"] >= 20].copy()
display(summary_filtered.sort_values("ack_success_rate", ascending=False))


# Největší latency outliers
outliers_latency = df.nlargest(10, "latency_send_to_ack_s")[[
    "dst_chain_guess", "send_timestamp", "recv_timestamp", "ack_timestamp",
    "latency_send_to_recv_s", "latency_send_to_ack_s", "latency_recv_to_ack_s",
    "has_recv", "has_ack"
]]

display(outliers_latency)


# Největší cost outliers
outliers_cost = df.nlargest(10, "network_cost_raw")[[
    "dst_chain_guess",
    "send_fee_amount", "send_fee_denom",
    "recv_fee_amount", "recv_fee_denom",
    "ack_fee_amount", "ack_fee_denom",
    "user_cost_raw", "relayer_cost_raw", "network_cost_raw",
    "has_recv", "has_ack"
]]

display(outliers_cost)
```
